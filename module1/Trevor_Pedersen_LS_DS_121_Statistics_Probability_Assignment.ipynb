```python
### YOUR WORK HERE
import pandas as pd
import numpy as np
from scipy import stats
```

```python
#read in the data
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data')
df
```

```python
#fix column headers
column_headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration",
"num-of-doors", "body-style", "drive-wheels", "engine-location",
"wheel-base", "length", "width", "height", "curb-weight",
"engine-type", "num-of-cylinders", "engine-size", "fuel-system",
"bore", "stroke", "compression-ratio", "hoursepower", "peak-rpm",
"city-mpg", "highway-mpg", "price"]

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data', names=column_headers)

df
```

### 2) Pretend that this dataset represents the cars at a used car lot in your local town. 

- Use df.sample() to pick a random sample of 10 cars. Note that because this sample is random we are going to set the `random_state` so that all of us in the class get the same random sample. Please set your random state to `30` when using `df.sample()`

Is your sample reflective of the population value in regards to highway-mpg? Find the mean for `highway-mpg` for the entire dataset and use a 1-sample t-test to compare your estaimated sample mean to the population mean. Can you say that they are different? 

```python
### YOUR WORK HERE

#Prevent scientific notation number format
pd.options.display.float_format = '{:.0f}'.format

#Sample
sample = df.sample(n=10, random_state=30)
print(sample)
sample.mean()
```

```python
#We are getting some obscenely large numbers, so we must be doing something wrong
#Not entirely sure why, but I suspect it might have something to do with the datatypes so I'm going to check that
sample.dtypes
```

```python
#Yup, if we look at price it is a object
#Actually, if we look at any of our messed up rows we will see that they are all string objects
#Lets convert all strings into numbers

#iterate through every column, check if it is a object, if true convert to float, if errors thrown do nothing on column
for column in df:
  if df[column].dtype == 'object':
    df[column] = df[column].astype(float, errors='ignore')

df.dtypes
```

```python
#Okay, lets finally rerun the cells above

#Sample
sample = df.sample(n=10, random_state=30)
sample.mean()
```

```python
#Okay, this looks a lot more normal
```

```python
#Look at overall mean
pop_mean = df['highway-mpg'].mean()
pop_mean
```

```python
#perform a 1sample ttest
stats.ttest_1samp(sample['highway-mpg'], popmean=pop_mean)
```

```python
#A t-value of 0 indicates that the sample is exactly equal with the null hypothesis
#We have a value greater than 0, so that means our sample's mean is greater than the population mean
```

The salesman says the cars he sells typically have a fuel efficiency of about 35 miles per gallon on the highway. You want to verify his claim but can only test 10 cars. Using your sample of 10, test his claim and report your results.

```python
### YOUR WORK HERE
sample['highway-mpg'].mean()
```

```python
#Our mean suggests the salesman is overstating MPG by 2.2
```

Imagine that you now have the capacity to test 100 cars on the lot. Using the same random state of `30`, take a sample of 100 cars. Run a t-test to verify the salesman's claim with your new larger sample. Do you reach the same conclusion as you did with the sample size of 10? 

```python
### YOUR WORK HERE
sample = df.sample(n=100, random_state=30)
sample['highway-mpg'].mean()
```

```python
#Now the salesman is overstating by 5 MPG!
```

```python
stats.ttest_1samp(sample['highway-mpg'], popmean=35)
```

Why might these two t-tests using the same dataset lead to different conclusions about the salesman's claim?

Because we are taking a sample out of the total population. Depending on what values we happen to get we can get wildly different answers. Our first sample of 10 is fairly small, so it can lead to huge variances in values. Our second sample is much larger, so the chance of variance from the population is much lower.

## Practice 2-Sample T-tests

Two Sample t-tests determine whether or not two sample means are statistically different from each other. 

This portion of your assignment requires you to determine which issues have "statistically significant" differences between political parties in this [1980s congressional voting data](https://archive.ics.uci.edu/ml/datasets/Congressional+Voting+Records). 


### 1) Load the data

The data consists of 435 instances (one for each congressperson), a class (democrat or republican), and 16 binary attributes (yes or no for voting for or against certain issues). Be aware - there are missing values!

- Read the dataset in from UCI, you'll need to provide a list of column headers
- Encode "yes" votes as 1 and "no" votes as 0. (You can use `df.replace()` to swap out these values)
- Use dataframe filtering to split the dataframe into two new dataframes based on party. Hold all republicans in one dataframe and all democrats in the other. These will be our two different "samples."



```python
### YOUR WORK HERE
#read in data
headers = ['class-name', 'handicapped-infants', 'water-project-cost-sharing', 'adoption-of-the-budget-resolution', 'physician-fee-freeze', 'el-salvador-aid',
           'religious-groups-in-schools', 'anti-satellite-test-ban', 'aid-to-nacaraguan-contras', 'mx-missle', 'immigration', 'synfuels-corporation-cutback',
           'education-spending', 'superfund-right-to-sue', 'crime', 'duty-free-exports', 'export-administration-act-south-africa']

df2 = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data', names=headers)
df2
```

```python
df2.dtypes
```

```python
#replace values
df2 = df2.replace('y', 1)
df2 = df2.replace('n', 0)
df2 = df2.replace('?', np.NaN)
df2
```

```python
#group by democrat and republican
sample_dem = df2[df2['class-name'] == 'democrat']
sample_repub = df2[df2['class-name'] == 'republican']

print(sample_dem)
sample_repub
```

### 2) Perform two sample T-tests on different issues and report the results.

- Find an issue that democrats support more than republicans with p < 0.01 (significant at the 99% level).
- Find an issue that republicans support more than democrats with p < 0.01 (significant at the 99% level).
- Find an issue where the difference between republicans and democrats has p > 0.1 (Not significant at the 90% level - i.e. there may not be much of a difference the two sample means)

Please for each test that you run state your null and alternative hypothesis and then write a conclusion reflecting on the null and alternative hypothesis.

Remember, that two-sample t-tests will only tell us if the two groups are *different* from one another. We'll have to look at their sample means directly or use the sign on the t-statistic to know which of the two sample means is larger. 

```python
### YOUR WORK HERE
for column in df2:
  if column == 'class-name':
    continue
  data = stats.ttest_ind(sample_dem[column], sample_repub[column], nan_policy='omit')
  

  if data[0] > 0 and data[1] < .01:
    #repub support
    print('Conclusion: Based on a t-statistic of:', data[0], 'and a p-value of:', data[1], 'I reject the null hypothesis that democrats support', column, 'at the same rate as republicans')
    pass
  pass
```

```python
for column in df2:
  if column == 'class-name':
    continue
  data = stats.ttest_ind(sample_dem[column], sample_repub[column], nan_policy='omit')
  #dem support
  if data[0] < 0 and data[1] < .01:
    print('Conclusion: Based on a t-statistic of:', data[0], 'and a p-value of:', data[1], 'I reject the null hypothesis that republicans support', column, 'at the same rate as democrats')
    pass
  pass
```

```python
for column in df2:
  if column == 'class-name':
    continue
  data = stats.ttest_ind(sample_dem[column], sample_repub[column], nan_policy='omit')
  
  #equal support
  if data[1] > .1:
    print('Conclusion: Based on a t-statistic of:', data[0], 'and a p-value of:', data[1], 'I fail to reject the null hypothesis that democrats and republicans support', column, 'at the same rate')
    pass
  pass
```
